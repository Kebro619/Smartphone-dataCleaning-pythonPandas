```python
def fast_charging_extractor(item):
    if type(item) == list:
        if len(item) == 1:
            return item[0]
        else:
            return 0
    else:
        return -1
```

```python
df1['fast_charging'] = df1['fast_charging'].apply(fast_charging_extractor).astype(int)
```

```python
df1.head()
```

```python
# making fast_charging_available colmn
def fast_charging_status(num):
    if num >=0:
        return 1
    else :
        return -1
```

```python
df1.head()
```

```python
def fast_charging_convert(num):
    if num > 0:
        return num
    elif num == 0:
        return np.nan
    else:
        return 0
```

```python
df1['fast_charging'] = df1['fast_charging'].apply(fast_charging_convert)
```

```python
def os_extractor(text):
  if pd.isna(text):
    return 'unknown'
  elif 'Android' in text:
    return 'android'
  elif 'iOS' in text:
    return 'ios'
  elif 'Harmony' in text or 'Hongmeng' in text or 'EMUI' in text:
    return 'other'
  else:
    pass
```
