## Part2 -Data Transformation

### making two column out of model (brand_name, model)

```python
df1.shape
```

```python
brand_names = df1['model'].str.split().str.get(0)
brand_names
```

```python
df1.insert(0,'brand_names',brand_names)
```

```python
df1['brand_names'] =df1['brand_names'].str.lower()
```

```python
df1.head()
```

### making three new columns out of sim column (has_5g,has_nfc,has_irblaster)

```python
has_5g = df1['sim'].str.contains('5G')
has_nfc = df1['sim'].str.contains('NFC')
has_ir_blaster = df1['sim'].str.contains('IR Blaster')
```

```python
df1.insert(5,'has_5g',has_5g)
df1.insert(6,'has_nfc',has_nfc)
df1.insert(7,'has_ir_blaster',has_ir_blaster)
```

```python
df1.head()
```

### making 3 news columns out of processor column(processor_name,num_core,processor_speed)

```python
df1['processor']
```

```python
processor_name = df1['processor'].str.split(',').str.get(0)
num_core = df1['processor'].str.split(',').str.get(1)
processor_speed = df1['processor'].str.split(',').str.get(2)
```

```python
df1.insert(9,'processor_name',processor_name)
df1.insert(10,'num_core',num_core)
df1.insert(11,'processor_speed',processor_speed)
```

```python
df1.head()
```
